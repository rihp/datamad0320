```python
df = pd.read_sql_query("""
SELECT 
    titleauthor.title_id,
    titleauthor.au_id,
    
    titles.price *
        sales.qty *
        titles.royalty/100 *
        titleauthor.royaltyper/100
            AS `SalesRoyalty`
    
FROM titleauthor

JOIN titles
    ON titleauthor.title_id=titles.title_id
JOIN sales
    ON titleauthor.title_id=sales.title_id

ORDER BY title_id
""", engine)
df
```

# Step 3

```python
df = pd.read_sql_query("""
SELECT 
  #  titleauthor.title_id,
  #  titleauthor.au_id,
    
    titles.title_id,
    sum(titles.price *
        sales.qty *
        titles.royalty/100 *
        titleauthor.royaltyper/100
        )  AS `SalesRoyalty`,
    
    titles.advance,
    
    `SalesRoyalty` + titles.advance
    
FROM titleauthor

JOIN titles
    ON titleauthor.title_id=titles.title_id
JOIN sales
    ON titleauthor.title_id=sales.title_id
    

GROUP BY titles.title_id

ORDER BY titles.title_id
""", engine)
df
```
