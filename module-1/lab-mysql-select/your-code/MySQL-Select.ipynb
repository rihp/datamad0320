# Challenge 1: Who Have Published What At Where?

```python
df = pd.read_sql_query("""
SELECT  
    authors.au_id "AUTHOR ID",
    authors.au_lname "LAST NAME",
    authors.au_fname "FIRST NAME",
    titles.title "TITLE",
    publishers.pub_name "PUBLISHER"
FROM titleauthor
LEFT JOIN authors
    ON titleauthor.au_id=authors.au_id
LEFT JOIN titles
    ON titleauthor.title_id=titles.title_id
LEFT JOIN publishers
    ON titles.pub_id=publishers.pub_id 
""", engine)

df
```

# Challenge 2 Who Have Published How Many At Where?

```python
df = pd.read_sql_query("""SELECT  
	authors.au_id "AUTHOR ID",
	authors.au_lname "LAST NAME",
    authors.au_fname "FIRST NAME",
    publishers.pub_name "PUBLISHER",
	count(publishers.pub_id) `TITLE COUNT`
FROM titleauthor
LEFT JOIN authors
	ON titleauthor.au_id=authors.au_id
LEFT JOIN titles
	ON titleauthor.title_id=titles.title_id
LEFT JOIN publishers
	ON titles.pub_id=publishers.pub_id 
GROUP BY publishers.pub_name, authors.au_id
ORDER BY `TITLE COUNT` DESC;
""", engine)
df
```

```python
df = pd.read_sql_query("""SELECT  
    authors.au_id "AUTHOR ID",
    authors.au_lname "LAST NAME",
    authors.au_fname "FIRST NAME",
    sum(sales.qty) `TOTAL`
FROM titleauthor
LEFT JOIN authors
    ON titleauthor.au_id=authors.au_id
LEFT JOIN titles
    ON titleauthor.title_id=titles.title_id
INNER JOIN sales
    ON titleauthor.title_id=sales.title_id
GROUP BY titleauthor.au_id
ORDER BY `TOTAL` DESC LIMIT 3; 

""", engine)
df
```

```python
# EN ESTE EJERCICIO NO HE LOGRADO QUE EL QUERY INCLUYA
# A LOS AUTORES QUE NO HAN TENIDO VENTAS.

df = pd.read_sql_query("""
SELECT 
    authors.au_id "AUTHOR ID",
    authors.au_lname "LAST NAME",
    authors.au_fname "FIRST NAME",
    IFNULL(sum(sales.qty),0) "SALES"
  #   sales.qty "SALES"
FROM titleauthor
LEFT JOIN authors
    ON titleauthor.au_id=authors.au_id
LEFT JOIN titles
    ON titleauthor.title_id=titles.title_id
LEFT JOIN sales
    ON titleauthor.title_id=sales.title_id
GROUP BY titleauthor.au_id #, sales.qty
ORDER BY SALES ASC; 
""", engine)
df
```
